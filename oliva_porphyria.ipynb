### This demonstrates the Python translation of the "oliva" demo program written in BASIC included with [The Algorithmic Beauty of Sea Shells](https://www.springer.com/gb/book/9783540921417) by [Hans Meinhardt](http://dev.biologists.org/content/develop/early/2016/03/22/dev.137414.full.pdf). It simulates an activator-inhibitor model for the pattern formation on the shells of [Oliva](https://en.wikipedia.org/wiki/Oliva_porphyria) [porphyria](http://www.marinespecies.org/aphia.php?p=taxdetails&id=208377).
(See also: [Meinhardt 1987](http://rocs.hu-berlin.de/complex_sys_2017/resources/Papers/Meinhardt_1987.pdf))

```python
from itertools import islice

import matplotlib.pyplot as plt
import numpy as np

from oliva import Oliva
```

A numpy array of shape `(x-pixels,2,y-pixels)` is returned, the 0th substance is the activator, the 1st is the inhibitor.

```python
shell = Oliva(width=1200)
cells = np.stack(islice(shell,0,1200))
cells.shape
```

The activator concentration is plotted using a threshold of 0.5.

```python
%matplotlib inline
plt.figure(figsize=(12,12))
plt.gray()
plt.imshow(cells[:,0,:] < 0.5)
```
